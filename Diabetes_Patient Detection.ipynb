```python
#getting 2D Tabular data for the given dataset
#returning it's first 5 rows

data=pd.DataFrame(data=data)
data.head()
```

```python
#using train_test_split function to split data into training and testing data.
#using data.drop() method to remove Outcome at index 1
#test set size is set to 0.20, train size will be 0.80 by default
#random state is set to 150 to shuffle the data between training and testing data

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.drop('Outcome',axis=1),data['Outcome'],test_size=0.20,random_state=150)
```

```python
#returning first 5 rows of testing data set for the first sequence

X_test.head()
```

```python
#importing LogisticRegression function from sklearn.model_selection package
#using Ibfgs algorithm to solve multiclass problem it handles multinomial loss.
#maximum number of iterations for solvers to converge is set to 7600.
#using fit method by using training data as an input to train the machine learning model 

from sklearn.linear_model import LogisticRegression
LRModel=LogisticRegression(solver='lbfgs', max_iter=7600)
LRModel.fit(X_train,y_train)
```

```python
#predicting output for each instance in Test data
predictions_diabetes=LRModel.predict(X_test)
```

```python
#importing classification report function to get precision, recall and f1-score on per class basis
#importing confusion matrix to show how many prediction are correct and incorrect per class
#printing classification report between true values of data and predicted output of model

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions_diabetes))
```
